```python
import sys
import pandas as pd
import numpy as np

sys.path.append('.')
data_daily = pd.read_csv('./data/48_Industry_Portfolios_Daily.csv', index_col=0)

# Trim the data
data_daily = data_daily.iloc[('197401' <= data_daily.index.values.astype(str)) & (data_daily.index.values.astype(str) <= '202012')]

# Convert the index to datetime
data_daily.index = pd.to_datetime(data_daily.index, format='%Y%m%d')
daily_date = data_daily.index.date

# The rate of return matrix, fill the missing values with 100%
Return = data_daily.values.T.astype(float)
Return[Return < -99]
Gross_return = (Return + 100) / 100 # Gross Return

print(Gross_return.shape)

# Check the first few rows of the dataset
data_daily.head()

window_size = 120
```

```python
# Check the first few rows of the dataset
data_daily.head()
```

```python
print(Gross_return)

# taks the 120 columns from Gross_return starting from the 1200th column
# data = Gross_return[:, 1200:1200 + window_size]
data = Gross_return[:, :window_size]
# Check the shape of the new dataset
print(data.shape)
```

```python
# convariance matrix
cov_matrix = np.cov(data)
# Check the shape of the covariance matrix
print(cov_matrix.shape)
# Check the first few rows of the covariance matrix
print(cov_matrix[:5, :5])
```

```python
# Eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
# Check the shape of the eigenvalues and eigenvectors
print(eigenvalues.shape)
print(eigenvectors.shape)
# Check the first few rows of the eigenvalues and eigenvectors
print(eigenvalues[:48])
print(eigenvectors[:5, :48])
```

```python
sorted_indices = np.argsort(eigenvalues)[::-1]
print(sorted_indices)
# Sort the eigenvalues and eigenvectors
sorted_eigenvalues = eigenvalues[sorted_indices]
sorted_eigenvectors = eigenvectors[:, sorted_indices]
print(sorted_eigenvalues.shape)
print(sorted_eigenvectors.shape)

print(eigenvectors[:5, :48])
print(sorted_eigenvectors[:5, :48])
```
